```python
import requests
import pandas as pd
```

```python
# Fetch dados da api
url = "https://dados.mobilidade.rio/gps/brt"

response = requests.get(url)   
data = response.json()
df = pd.DataFrame(data['veiculos'])
```

```python
# Dando uma olhada na estrutura e dados
df.head(15)
```

```python
# Mais info do df
df.info()
```

```python
# Verificando se temos nulo
df.isnull().sum()
```

```python
# Retornando Lat e Long range buscando por inconsistencias
print('Latiture range:', df['latitude'].min(), "to", df['latitude'].max())
print('Longitude range:', df['longitude'].min(), 'to', df['longitude'].max())
```

```python
# Retornando descricao de velocidade buscando por inconsistencias
df['velocidade'].describe()
```
